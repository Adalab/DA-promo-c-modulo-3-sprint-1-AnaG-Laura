```python
# Importamos las librerias necesarias
import pandas as pd

from sklearn.preprocessing import OneHotEncoder  

import warnings
warnings.filterwarnings('ignore')
```

```python
# abrimos nuestro DataFrame
df = pd.read_csv("archive/anova.csv", index_col = 0)
df.head()
```

```python
df = df.reset_index()
```

```python
df.drop('country', axis=1, inplace=True)
```

```python
df.head(2)
```

Utilizaremos nuestra variable 'region', la cual es nominal, puesto que no tiene orden.

```python
df['region'].unique()
```

Podriamos utilizar tanto el método OneHotEncoder o el método get_dummies.

```python
df_steam_cat = df.select_dtypes(include="O")
```

```python
df_steam_cat.head()
```

```python
df_steam_cat.region.unique()
```

```python
dummies = pd.get_dummies(df_steam_cat["region"], prefix_sep = "_", prefix = "region", dtype = int)
dummies.head(2)
```

```python
df_dummies = pd.concat([df_steam_cat, dummies], axis = 1)
df_dummies.head(2)
```

```python
final = pd.concat([df,df_dummies],axis=1)
```

```python
final.head(5)
```

```python
df.drop('region', axis=1, inplace=True)
```

```python
final.isnull().sum()
```

```python
final.to_csv('archive/encoding.csv', index=False)
```
