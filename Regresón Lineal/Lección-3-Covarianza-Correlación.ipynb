#### 1. Pregunta teórica: ¿Qué diferencia existe entre la covarianza y la correlación?

La covarianza indica si ambas variables varían en la misma dirección o en dirección opuesta (positiva o negativa), 
mientras que la correlación explica el cambio de proporción de una variable con otra.

#### 2. Calculad la covarianza y la correlación de nuestro dataset. No nos vale solo con que la calculéis, debéis hacer una interpretación de los resultados obtenidos. Calculad solo la correlación de Pearson.


- Covarianza

```python
# la representamos visualmente
sns.pairplot(df, palette='flare');
```

```python
# lo representamos analíticamente
df.cov()
```

Podemos ver que tenemos alguna covarianza positiva y alguna negativa. Por ejemplo cuando sube la columna de Economia baja el happiness rank, y es que realmente lo que nos interesa es tener un happiness rank bajo, ya que cuanto mas bajo mas felices son las personas que viven alli. Pasa igual con el resto de valores comparandolos con happiness rank. Sin embargo comparandolos con happiness score casi todas las variables tienen covaarianza positiva, cuanto mas aunmenta el valor de las variables mas aumenta la suma de felicidad.

- Correlación

```python
corr = df.corr()
corr
```

#### 3. Con los resultados de la correlación del ejercicio anterior, cread un heatmap.


#### 4. Guardad la gráfica del heatmap.

```python
plt.figure(figsize = (10,10))

sns.heatmap(corr, 
            cmap = "flare", 
           annot = True)

plt.savefig("archive/heatmap.png");
```

Con la ayuda del Heatmap, vemos que las variables predictivas que más correlación tienen con nuestra variable respuesta són la economía, la familia y la salud (esperanza de vida). Seguidas de la libertad, y la distopia residual. Y las que prácticamente no tienen ningúna correlación son la generosidad y la confianza en el gobierno.
